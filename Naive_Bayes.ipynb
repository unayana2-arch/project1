```python
ip=df.iloc[:,0:4]
op=df['species']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(ip,op,random_state=20)
```

```python
from sklearn.naive_bayes import GaussianNB,BernoulliNB,MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
l=[GaussianNB(),BernoulliNB(),MultinomialNB()]
for i in l:
  model=i
  model.fit(x_train,y_train)
  pred=model.predict(x_test)
  print(i,'= ',accuracy_score(y_test,pred))
```

```python
from sklearn.model_selection import cross_val_score
score=cross_val_score(MultinomialNB(),x_train,y_train)
print(score.mean(),score.std())
```

```python
from sklearn.model_selection import GridSearchCV
param=[
    {'alpha':[1.0], 'force_alpha':[True], 'fit_prior':[True], 'class_prior':[None]}
]
grid=GridSearchCV(MultinomialNB(),param,cv=3,verbose=1)
```

```python
grid.fit(x_train,y_train)
grid.best_params_
```
